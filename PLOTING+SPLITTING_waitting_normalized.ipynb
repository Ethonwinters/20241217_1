```python
import torch
import matplotlib.pyplot as plt
import numpy as np

A = torch.tensor([[-0.0221, -0.0238, -0.0040, -0.0359,  0.0121, -0.0363,  0.0179,
                   -0.0455, -0.0290,  0.0146,  0.0790,  0.0627,  0.0028, -0.0297,
                   -0.0365, -0.0120,  0.0167,  0.0238, -0.0127, -0.0220],
                  [-0.0528, -0.0106, -0.0692, -0.0411, -0.1033, -0.1809, -0.0551,
                   0.0129, -0.0379, -0.0199, -0.0544, -0.0761, -0.0917, -0.0045,
                   0.0092, -0.0732, -0.0230, -0.0558, -0.0678, -0.0435]], device='cuda:0')

# Step 1: Flatten the tensor to 1x400
flattened_A = A.flatten()

# Step 2: Get the values that are out of the range [-0.12, 0.12]
out_of_range_A = flattened_A[(flattened_A < -0.12) | (flattened_A > 0.12)]

# Step 3: Randomly remove 80% of the out-of-range values
num_to_remove = int(len(out_of_range_A) * 0.8)
indices_to_remove = np.random.choice(len(out_of_range_A), num_to_remove, replace=False)

# Remove the randomly selected values from out_of_range_A
out_of_range_A = torch.tensor(np.delete(out_of_range_A.cpu().numpy(), indices_to_remove))

# Step 4: Get the values that are within the range [-0.12, 0.12]
in_range_A = flattened_A[(flattened_A >= -0.12) & (flattened_A <= 0.12)]

# Step 5: Combine the in-range values with the remaining out-of-range values
filtered_A = torch.cat((in_range_A.cpu(), out_of_range_A))

# Step 6: Visualize the filtered values as a scatter plot
filtered_A_numpy = filtered_A.cpu().numpy()

# Create a figure and axis
plt.figure(figsize=(10, 6))

# Add shading for the region between -0.12 and 0.12
plt.axhspan(-0.12, 0.12, color='green', alpha=0.1, label='Shaded Area (-0.12 to 0.12)')

# Scatter plot for filtered values
plt.scatter(range(len(filtered_A_numpy)), filtered_A_numpy, color='red', marker='x', alpha=1, label='Filtered Points')

# Add horizontal lines at y = -0.12 and y = 0.12
plt.axhline(y=-0.12, color='brown', linestyle='--', label='y = -0.12')
plt.axhline(y=0.12, color='brown', linestyle='--', label='y = 0.12')

# Add text labels for the horizontal lines
plt.text(0, -0.12, 'y = -0.12', color='r', verticalalignment='bottom', horizontalalignment='left')
plt.text(0, 0.12, 'y = 0.12', color='g', verticalalignment='top', horizontalalignment='left')

# Label the axes and add a title
plt.xlabel('Index')
plt.ylabel('Error Value')
plt.title('Scatter Plot of Filtered 20x20 Error Tensor')

# Show grid, legend, and plot
plt.grid(True)
plt.legend()
plt.show()
```

```python
import os
import shutil
import random

# 
data_dir = r'C:\Users\ROG\data_all_wz'
input_dir = data_dir  #  (input0-input299)
target_dir = data_dir  #  (target0-target299)

# 
test_ratio = 0.1  # 10%
val_ratio = 0.1   # 10%

# 
input_files = [f for f in os.listdir(input_dir) if f.startswith('input') and f.endswith('.pt')]
target_files = [f for f in os.listdir(target_dir) if f.startswith('target') and f.endswith('.pt')]

# 
print(":", input_files)
print(":", target_files)

# inputtarget
assert len(input_files) == len(target_files), "input  target "

# 
data_pairs = list(zip(input_files, target_files))
random.shuffle(data_pairs)

# 
total_files = len(data_pairs)
test_size = int(total_files * test_ratio)
val_size = int(total_files * val_ratio)

# 
test_data = data_pairs[:test_size]
val_data = data_pairs[test_size:test_size + val_size]
train_data = data_pairs[test_size + val_size:]

# 
def create_directory(dir_name):
    if not os.path.exists(dir_name):
        os.makedirs(dir_name)

# : train, val, test
create_directory(os.path.join(data_dir, 'train', 'inputs'))
create_directory(os.path.join(data_dir, 'train', 'targets'))
create_directory(os.path.join(data_dir, 'val', 'inputs'))
create_directory(os.path.join(data_dir, 'val', 'targets'))
create_directory(os.path.join(data_dir, 'test', 'inputs'))
create_directory(os.path.join(data_dir, 'test', 'targets'))

# 
def move_files(data, subset):
    for input_file, target_file in data:
        input_src = os.path.join(input_dir, input_file)
        target_src = os.path.join(target_dir, target_file)
        
        input_dest = os.path.join(data_dir, subset, 'inputs', input_file)
        target_dest = os.path.join(data_dir, subset, 'targets', target_file)
        
        shutil.copy(input_src, input_dest)
        shutil.copy(target_src, target_dest)

# 
move_files(train_data, 'train')
move_files(val_data, 'val')
move_files(test_data, 'test')

print(f"\n{len(train_data)}\n{len(val_data)}\n{len(test_data)}")
print(f"\n{len(train_data)}\n{len(val_data)}\n{len(test_data)}")
```

```python
import os
import re

# 
test_dir = r'C:\Users\ROG\data_all_wz\test'  # 

#  .pt 
pt_files = [f for f in os.listdir(test_dir) if f.endswith('.pt')]

# 
def get_file_number(filename):
    #  input###.pt  target###.pt
    match = re.search(r'(\d+)', filename)
    return int(match.group(1)) if match else -1

#  input  target 
input_files = [f for f in pt_files if f.startswith('input')]
target_files = [f for f in pt_files if f.startswith('target')]

# 
sorted_input_files = sorted(input_files, key=get_file_number)
sorted_target_files = sorted(target_files, key=get_file_number)

#  input  target 
assert len(sorted_input_files) == len(sorted_target_files), "input  target "

# 
for i, (input_file, target_file) in enumerate(zip(sorted_input_files, sorted_target_files), 1):
    #  input1.pt, target1.pt
    new_input_filename = f"input{i}.pt"
    new_target_filename = f"target{i}.pt"
    
    # 
    old_input_file_path = os.path.join(test_dir, input_file)
    old_target_file_path = os.path.join(test_dir, target_file)
    
    # 
    new_input_file_path = os.path.join(test_dir, new_input_filename)
    new_target_file_path = os.path.join(test_dir, new_target_filename)
    
    # 
    os.rename(old_input_file_path, new_input_file_path)
    os.rename(old_target_file_path, new_target_file_path)
    
    print(f"Renamed {input_file} to {new_input_filename}")
    print(f"Renamed {target_file} to {new_target_filename}")

print("")
```

```python
import os
import torch
import re

# 
dir1 = r'C:\Users\ROG\data_all_wz\test'         # 
dir2 = r'C:\Users\ROG\data_all_wz\test_original' # 

#  input  target 
def get_sorted_files(directory, prefix):
    files = [f for f in os.listdir(directory) if f.startswith(prefix) and f.endswith('.pt')]
    
    # 
    def get_file_number(filename):
        match = re.search(r'(\d+)', filename)
        return int(match.group(1)) if match else -1
    
    # 
    sorted_files = sorted(files, key=get_file_number)
    return sorted_files

#  input  target 
input_files_dir1 = get_sorted_files(dir1, 'input')
target_files_dir1 = get_sorted_files(dir1, 'target')
input_files_dir2 = get_sorted_files(dir2, 'input')
target_files_dir2 = get_sorted_files(dir2, 'target')

# 
for i in range(len(input_files_dir1)):
    #  input  target 
    input_file1 = input_files_dir1[i]
    target_file1 = target_files_dir1[i]
    
    input_file2 = input_files_dir2[i]
    target_file2 = target_files_dir2[i]
    
    # 
    input_tensor1 = torch.load(os.path.join(dir1, input_file1))
    target_tensor1 = torch.load(os.path.join(dir1, target_file1))
    
    input_tensor2 = torch.load(os.path.join(dir2, input_file2))
    target_tensor2 = torch.load(os.path.join(dir2, target_file2))
    
    # 
    input_equal = torch.equal(input_tensor1, input_tensor2)
    target_equal = torch.equal(target_tensor1, target_tensor2)
    
    # 
    if input_equal and target_equal:
        print(f"{input_file1}  {input_file2} ")
        print(f"{target_file1}  {target_file2} ")
    else:
        if not input_equal:
            print(f"{input_file1}  {input_file2} ")
        if not target_equal:
            print(f"{target_file1}  {target_file2} ")

print("")
```
